```python
import numpy as np
import pandas as pd
import datetime
```

```python
# Reading Output template file - check file name, format, column headers, spellings etc
output = pd.read_excel("C:/Users/TE335168/Desktop/MIS Team/Adhoc/PCN report/MAIN SOURCE/PCN Report Template.xlsx")

# For reading PCN source file - check file name, format, column headers, spellings etc
file = pd.ExcelFile("C:/Users/TE335168/Desktop/MIS Team/Adhoc/PCN report/MAIN SOURCE/FY 2022/Oct 2022/Source.xlsx")
source = pd.read_excel(file, 'Export Worksheet')
discrete = pd.read_excel(file, 'Discrete EOL pns')

# Reading Restricted Part Info file
rest = pd.read_excel("C:/Users/TE335168/Desktop/MIS Team/Adhoc/PCN report/MAIN SOURCE/FY 2022/Oct 2022/Restricted.xlsx")

# Verifying source and output
print('Output')
display(output.head())
print("\n")
print('Source')
display(source.head(5))

# Dropping unnecessary cols, renaming cols
source = source.drop(['PCN_TYPE','PCN LT'], axis = 1)
source.rename(columns={'PCN Type': 'PCN_Type'}, inplace=True)

print("\n")
print('Updated Source')
source.head(5)
```

```python
# Filtering df based on PCN Type = EOL
eol = source.loc[(source['PCN_Type'] == 'EOL')]
#print('eol filtered')
eol = eol[pd.notnull(eol['PART_NUMBER'])]
#display(eol.head(5))

# Getting unique list of part_numbers
prtnbr = eol['PART_NUMBER'].dropna().unique().tolist()
print("Unique part number count: ",len(prtnbr))

# Dropping duplicate values in filtered data keeping first value and resetting index
eol = eol.drop_duplicates(subset='PART_NUMBER', keep="first")
eol.reset_index(drop=True, inplace=True)
print("eol shape: ",eol.shape)
print("\n")
eol.head()
```

```python
#Updating values in output template
output['PART_NUMBER'] = eol['PART_NUMBER']
output['PCN Type'] = 'EOL'
output['Discontinued?'] = 'Yes'
output['PCN Number'] = eol['PCN_NUMBER']
output['Reason for discontinuance'] = eol['REASON_FOR_CHANGES']
output['Part Description'] = eol['GENERAL_PRODUCT_DESCRIPTION']
output['Last Order Date'] = eol['LAST_ORDER_DATE']
```

```python
# Verifying Discrete EOL Tab
print(discrete.shape)

# Check if number of rows for unique part numbers = number of rows in Discrete EOL PNs sheet
# Updating column names
discrete.rename(columns={'P/N': 'PART_NUMBER', 'Replacement Part': 'Replacement', 'PCN LT': 'PCN_LT'}, inplace=True)
display(discrete.head())
```

```python
# Updating output template
#output['MOQ'] = discrete['MOQ'] check if MOQ is present in Discrete
#output['GPL'] = discrete['GPL'] check if GPL is present in Discrete
output['BU'] = discrete['BU']
output['Replacement Part'] = discrete['Replacement']
output['PCN LT'] = discrete['PCN_LT']
```

```python
# Input current date for calculating 'Can EOL be bought?'
date_entry = input('Enter a date in YYYY-MM-DD format')
year, month, day = map(int, date_entry.split('-'))
date = datetime.date(year, month, day)
date = pd.to_datetime(date)
date
#formatted_date = datetime.date.strftime(date1, "%m/%d/%Y")
```

```python
# Converting data type of Last Order Date col from string to datetime and removing //
output['Last Order Date'] = np.where(output['Last Order Date'] == '//', 0 , output['Last Order Date'])
output['Last Order Date'] = pd.to_datetime(output['Last Order Date'])
```

```python
# Appending status list based on conditions for 'Can EOL be bought?'
date1 = pd.to_datetime('1970-01-01')

status = []
for row in  output['Last Order Date']:
    if row > date:
        status.append('Yes')
    elif row == date1:
        status.append('Unknown')
    else:
        status.append('No')
```

```python
#Updating data in output template for replacement part and Last order Date in MM-DD-YYYY format
output['Can EOL be bought?'] = status
output['Replacement Part'] = np.where(output['Replacement Part'] == 0, 'N/A', output['Replacement Part'])
output['Last Order Date'] = output['Last Order Date'].dt.date
output['Last Order Date'] = pd.to_datetime(output['Last Order Date']).dt.strftime('%m/%d/%Y')
output['Last Order Date'] = np.where(output['Last Order Date'] == '01/01/1970', '-' , output['Last Order Date'])
```

```python
# For restricted part information
display(rest.head(2))
rest.shape
```

```python
rest = rest.drop(['Part Description','Part Status','Part Rev','RDO','Production Drawing','Project Nbr','Product Code'], axis=1)
rest['Restricted Part?'] = np.where(rest['Sales Intent'] == 'RESTRICTED SALES',"Yes","No")
rest.rename(columns={'TCPN': 'PART_NUMBER'}, inplace=True)
rest = rest.drop(['Sales Intent'],axis=1)
output = output.drop(['Restricted Part?'],axis=1)
output = pd.merge(output, rest, on ='PART_NUMBER', how ='left')
```
